# PROBLEM 3

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def run(n):
    
    prng = np.random.default_rng()
    a, b = 0, 1
    
# Get n random numbers from the exponential distribution
    x_is = prng.uniform(size=n)
    
    h_is = (b-a)*(np.log(x_is))*(x_is**3)*(np.exp(-x_is))*(1/6)
    
    estimate = np.sum(np.array(h_is))/n

    
    return estimate


def sim(N=[10,20,50,80,100,200,500,600,800,1000]):
    estimates=[]
    
    for k in N:
        
        a = run(k)
        estimates.append(a)
        
    plt.plot(N,estimates)
    

print(sim())
```

```python
import numpy as np
import math
from scipy.stats import norm, uniform

prng = np.random.default_rng()

nsamples = 10000

# Using an integral calculator
truth = 0.52479 

x_is = prng.uniform(low=0,high=1,size=nsamples)
g_of_xis = (np.exp(-x_is))/(1+(x_is**2))
h_of_xis = g_of_xis #np.where(x_is<1.0,0,g_of_xis)

estimate = np.sum(h_of_xis)/nsamples
std_error = math.sqrt(np.var(h_of_xis, ddof = 1) / nsamples)
rel_error = (np.abs(estimate - truth)/truth) * 100

print("True value ", truth)
print("Estimate : ", f"{estimate:e}", f"Standard Error {std_error:e}", \
f"Rel Error {rel_error:e}")
```
